```python
data = np.genfromtxt('sho.dat')
```

```python
nrows = 2
ncols = 1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, sharex=True, figsize=(ncols*14, nrows*7))

axes[0].plot(data[:,0], data[:,1], color='grey')
axes[1].plot(data[:,0], data[:,2], color='grey')

axes[0].scatter(data[:,0], data[:,1], color='red')
axes[1].scatter(data[:,0], data[:,2], color='green')

axes[0].set_ylabel('Deflection', fontsize=18)

axes[1].set_xlabel('Time [s]', fontsize=18)
axes[1].set_ylabel('Velocity', fontsize=18)

fig.tight_layout()
plt.show()
```

```python
nrows = 1
ncols = 1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*14, nrows*7))

axes.plot(data[:,1], data[:,2], color='grey')
axes.scatter(data[:,1], data[:,2], color='red')

axes.set_xlabel('Deflection', fontsize=18)
axes.set_ylabel('Velocity', fontsize=18)

plt.show()
```

```python
nrows = 3
ncols = 1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*14, nrows*7))

axes[0].plot(data[:,0], data[:,3], color='grey')
axes[1].plot(data[:,1], data[:,3], color='grey')
axes[2].plot(data[:,2], data[:,3], color='grey')

axes[0].scatter(data[:,0], data[:,3], color='red')
axes[1].scatter(data[:,1], data[:,3], color='green')
axes[2].scatter(data[:,2], data[:,3], color='blue')

axes[0].set_xlabel('Time [s]', fontsize=18)
axes[0].set_ylabel('Energy', fontsize=18)
axes[1].set_xlabel('Deflection', fontsize=18)
axes[1].set_ylabel('Energy', fontsize=18)
axes[2].set_xlabel('Velocity', fontsize=18)
axes[2].set_ylabel('Energy', fontsize=18)

plt.show()
```
