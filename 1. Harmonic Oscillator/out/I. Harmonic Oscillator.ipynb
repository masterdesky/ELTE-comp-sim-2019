# I. Harmonic Oscillator

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style(style='whitegrid')
```

```python
data = np.loadtxt('sho.txt')
```

```python
data[:,0]
```

```python
nrows = 2
ncols = 1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*14, nrows*7))

axes[0].scatter(data[:,0], data[:,1], color='red')
axes[1].scatter(data[:,0], data[:,2], color='green')

axes[0].set_xlabel('Time')
axes[0].set_ylabel('X coordinate')

axes[1].set_xlabel('Time')
axes[1].set_ylabel('Velocity')

fig.tight_layout()
plt.show()
```

```python
nrows = 1
ncols = 1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*14, nrows*7))

axes.scatter(data[:,1], data[:,2], color='red')

axes.set_xlabel('X coordinates')
axes.set_ylabel('Velocity')

plt.show()
```
